```python
# ==========================================
# AUTONOMOUS DELIVERY AGENT (My Project)
# ==========================================
# This is my mini project for 1st year BTech.
# Idea: A delivery bot in a city (grid) that finds path.
# I tried BFS, UCS (like Dijkstra), A* (smartest), and hill climbing (local search).
# Itâ€™s not perfect but works :)

import os, time, heapq
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import deque

# ----------------------------
# STEP 1: make folder for maps
# ----------------------------
os.makedirs("delivery_agent/maps", exist_ok=True)
print("ðŸ“‚ made project folders")
```

```python
# ----------------------------
# STEP 2: The Grid Environment
# ----------------------------
class Grid:
    def __init__(self, filename):
        # map file: first line has rows and cols
        with open(filename, "r") as f:
            lines = f.read().strip().split("\n")
        dims = lines[0].split()
        self.rows, self.cols = int(dims[0]), int(dims[1])
        self.grid = []
        self.start = None
        self.goal = None

        # reading each character in map
        for r, line in enumerate(lines[1:]):
            row = []
            for c, ch in enumerate(line):
                if ch == "S":   # start
                    self.start = (r, c)
                    row.append(1)
                elif ch == "G": # goal
                    self.goal = (r, c)
                    row.append(1)
                elif ch == "#": # wall
                    row.append(float("inf"))
                else:           # road
                    row.append(1)
            self.grid.append(row)
        self.grid = np.array(self.grid)

    def in_bounds(self, pos):
        r, c = pos
        return 0 <= r < self.rows and 0 <= c < self.cols

    def passable(self, pos):
        r, c = pos
        return self.grid[r][c] != float("inf")

    def neighbors(self, pos):
        r, c = pos
        # bot moves only up, down, left, right
        moves = [(1,0), (-1,0), (0,1), (0,-1)]
        out = []
        for dr, dc in moves:
            nxt = (r+dr, c+dc)
            if self.in_bounds(nxt) and self.passable(nxt):
                out.append(nxt)
        return out
```

```python
# ----------------------------
# STEP 3: BFS (brute force)
# ----------------------------
def bfs(grid):
    start, goal = grid.start, grid.goal
    q = deque([start])
    came = {start: None}

    while q:
        cur = q.popleft()
        if cur == goal:
            break
        for nxt in grid.neighbors(cur):
            if nxt not in came:
                q.append(nxt)
                came[nxt] = cur

    # rebuild path
    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = came.get(cur)
    path.reverse()
    return path
```

```python
# ----------------------------
# STEP 4: UCS (like Dijkstra)
# ----------------------------
def ucs(grid):
    start, goal = grid.start, grid.goal
    frontier = [(0, start)]
    came = {start: None}
    cost = {start: 0}

    while frontier:
        cur_cost, cur = heapq.heappop(frontier)
        if cur == goal:
            break
        for nxt in grid.neighbors(cur):
            new_c = cost[cur] + grid.grid[nxt]
            if nxt not in cost or new_c < cost[nxt]:
                cost[nxt] = new_c
                heapq.heappush(frontier, (new_c, nxt))
                came[nxt] = cur

    # rebuild path
    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = came.get(cur)
    path.reverse()
    return path, cost.get(goal, float("inf"))
```

```python
# ----------------------------
# STEP 5: A* (smartest one)
# ----------------------------
def h(a, b):
    # Manhattan distance
    return abs(a[0]-b[0]) + abs(a[1]-b[1])

def astar(grid):
    start, goal = grid.start, grid.goal
    frontier = [(0, start)]
    came = {start: None}
    cost = {start: 0}

    while frontier:
        _, cur = heapq.heappop(frontier)
        if cur == goal:
            break
        for nxt in grid.neighbors(cur):
            new_c = cost[cur] + grid.grid[nxt]
            if nxt not in cost or new_c < cost[nxt]:
                cost[nxt] = new_c
                pr = new_c + h(goal, nxt)
                heapq.heappush(frontier, (pr, nxt))
                came[nxt] = cur

    path = []
    cur = goal
    while cur is not None:
        path.append(cur)
        cur = came.get(cur)
    path.reverse()
    return path, cost.get(goal, float("inf"))
```

```python
# ----------------------------
# STEP 6: Hill climbing (local)
# ----------------------------
def hill_climb(grid, restarts=3):
    start, goal = grid.start, grid.goal
    for _ in range(restarts):
        cur = start
        path = [cur]
        while cur != goal:
            neigh = grid.neighbors(cur)
            if not neigh:
                break
            neigh.sort(key=lambda n: h(n, goal))
            nxt = neigh[0]
            if h(nxt, goal) >= h(cur, goal):
                break
            path.append(nxt)
            cur = nxt
        if cur == goal:
            return path
    return []
```

```python
# ----------------------------
# STEP 7: Maps
# ----------------------------
maps = {
    "small": """5 5
S....
.###.
..#..
.###.
....G
""",
    "medium": """8 8
S.......
.#####..
..#.....
..#..###
..#.....
..###..G
........
........
""",
    "large": """12 12
S...........
.#####......
....#.......
....#.......
....#.......
....#....###
....#.......
....#.......
....#.......
....#.......
....#.......
...........G
""",
    "dynamic": """6 6
S.....
.###..
..#...
.###..
....#.
.....G
"""
}

for name, content in maps.items():
    with open(f"delivery_agent/maps/{name}.map", "w") as f:
        f.write(content)
print("ðŸ—º maps saved")
```

```python
# ----------------------------
# STEP 8: Run Experiments
# ----------------------------
def run_exp(mapfile, planners):
    g = Grid(mapfile)
    out = []
    for name in planners:
        t0 = time.time()
        if name == "bfs":
            p = bfs(g)
            c = len(p)
        elif name == "ucs":
            p, c = ucs(g)
        elif name == "astar":
            p, c = astar(g)
        elif name == "local":
            p = hill_climb(g)
            c = len(p)
        else:
            continue
        t1 = time.time()-t0
        out.append({
            "Algo": name,
            "Map": mapfile.split("/")[-1],
            "PathLen": len(p),
            "Cost": c,
            "Time": round(t1,4)
        })
    return out

planners = ["bfs", "ucs", "astar", "local"]
all_res = []
for m in maps.keys():
    all_res.extend(run_exp(f"delivery_agent/maps/{m}.map", planners))

df = pd.DataFrame(all_res)
print("âœ… got results")
print(df)
```

```python
# ----------------------------
# STEP 9: Plot
# ----------------------------
def plot_metric(df, col, title):
    plt.figure(figsize=(9,5))
    for m in df["Map"].unique():
        sub = df[df["Map"]==m]
        plt.bar([f"{m}-{p}" for p in sub["Algo"]], sub[col])
    plt.title(title)
    plt.xticks(rotation=45)
    plt.ylabel(col)
    plt.show()

plot_metric(df, "PathLen", "Path Length")
plot_metric(df, "Cost", "Path Cost")
plot_metric(df, "Time", "Runtime")
```

```python
# ----------------------------
# STEP 10: Dynamic replanning
# ----------------------------
def demo_dynamic(mapfile):
    g = Grid(mapfile)
    print("\nFirst plan with A*:")
    p, c = astar(g)
    print("Path:", p, "Cost:", c)

    print("\nðŸš§ obstacle suddenly at (2,2)! replanning...")
    g.grid[2][2] = float("inf")
    new_p, new_c = astar(g)
    print("New path:", new_p, "Cost:", new_c)

demo_dynamic("delivery_agent/maps/dynamic.map")
```
